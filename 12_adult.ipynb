```python
melted = pd.melt(
    df,
    id_vars=['gender', 'race', 'native-country'],
    value_vars=['age', 'educational-num', 'hours-per-week'],
    var_name='attribute',
    value_name='value'
)

melted
```

```python
df_cast = melted.pivot_table(
    index=['gender', 'race', 'native-country'],
    columns='attribute',
    values='value',
    aggfunc='mean'
)

df_cast = df_cast.reset_index()  # Optional
df_cast.transpose()
```
